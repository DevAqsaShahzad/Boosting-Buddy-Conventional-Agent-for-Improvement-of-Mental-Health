```
with open('RomanUrduDatasetLabelled3Emotions.csv', 'r') as inp, open('myfile.txt', 'w') as out:
    for line in inp:
        line = line.replace(':', ':')
        out.write(line)
with open('myfile.txt') as f:
    df = f.read()
article_text = ''
for line in df:  
    article_text += line
    
corpus = nltk.sent_tokenize(article_text)
print(corpus)
```

```
for i in range(len(corpus )):
    corpus [i] = corpus [i].lower()
    corpus [i] = re.sub(r'\W',' ',corpus [i])
    corpus [i] = re.sub(r'\s+',' ',corpus [i])
```

```
wordfreq = {}
for sentence in corpus:
    tokens = nltk.word_tokenize(sentence)
    for token in tokens:
        if token not in wordfreq.keys():
            wordfreq[token] = 1
        else:
            wordfreq[token] += 1
```

```
import heapq
most_freq = heapq.nlargest(500, wordfreq, key=wordfreq.get)
```

```
sentence_vectors = []
for sentence in corpus:
    sentence_tokens = nltk.word_tokenize(sentence)
    sent_vec = []
    for token in most_freq:
        if token in sentence_tokens:
            sent_vec.append(1)
        else:
            sent_vec.append(0)
    sentence_vectors.append(sent_vec)
```

```
sentence_vectors = np.asarray(sentence_vectors)
sentence_vectors.shape
```

```
df=pd.DataFrame(sentence_vectors)
df.to_csv(r'BOWVec.csv', index= False)
df
```
