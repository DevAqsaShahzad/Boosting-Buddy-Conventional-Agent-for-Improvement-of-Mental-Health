```python
import pandas as pd
import numpy as np
import nltk
import string
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords, wordnet
from nltk.stem import WordNetLemmatizer
nltk.download('punkt')   
import random 
import bs4 as bs  
import urllib.request  
import re
```

```python
with open('C:/Users/Pisces Khan/Downloads/RomanUrduDatasetLabelled3Emotions.csv', 'r') as inp, open('myfile.txt', 'w') as out:
    for line in inp:
        line = line.replace(':', ':')
        out.write(line)
with open('myfile.txt') as f:
    df = f.read()
article_text = ''
for line in df:  
    article_text += line
    
corpus = nltk.sent_tokenize(article_text)
print(corpus)
```

```python
for i in range(len(corpus )):
    corpus [i] = corpus [i].lower()
    corpus [i] = re.sub(r'\W',' ',corpus [i])
    corpus [i] = re.sub(r'\s+',' ',corpus [i])
```

```python
print(len(corpus))
```

```python
print(corpus[30])
```

```python
wordfreq = {}
for sentence in corpus:
    tokens = nltk.word_tokenize(sentence)
    for token in tokens:
        if token not in wordfreq.keys():
            wordfreq[token] = 1
        else:
            wordfreq[token] += 1
            
```

```python
wordfreq
```

```python
import heapq
most_freq = heapq.nlargest(500, wordfreq, key=wordfreq.get)
```

```python
sentence_vectors = []
for sentence in corpus:
    sentence_tokens = nltk.word_tokenize(sentence)
    sent_vec = []
    for token in most_freq:
        sent_vec.append(token)
    for token in most_freq:
        if token in sentence_tokens:
            sent_vec.append(1)
        else:
            sent_vec.append(0)
    sentence_vectors.append(sent_vec)
```

```python
sentence_vectors = np.asarray(sentence_vectors)
sentence_vectors.shape
```

```python
df=pd.DataFrame(sentence_vectors)
df.to_csv(r'BOWVec.csv', index= False)
df
```

```python
bagofword = np.array(list(wordfreq.keys()))
bow  = pd.DataFrame(bagofword)


xd=pd.DataFrame(sentence_vectors)
d = [bow,xd]
df = pd.concat(d)
df.to_csv(r'BOWVec.csv', index= False)
bagofword
```
